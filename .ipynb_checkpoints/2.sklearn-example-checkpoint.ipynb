# 机器学习库sklearn

## sklearn简介

Scikit learn 也简称 sklearn, 是机器学习领域当中最知名的 python 模块之一.

Sklearn 包含了很多种机器学习的方式:

- Classification 分类
- Regression 回归
- Clustering 非监督分类
- Dimensionality reduction 数据降维
- Model Selection 模型选择
- Preprocessing 数据预处理

## 选择学习方法

进行机器学习，首先要选择适当的方法，来达到你的目标。

Sklearn 官网提供了一个流程图， 蓝色圆圈内是判断条件，绿色方框内是可以选择的算法：

从 START 开始，首先看数据的样本是否 >50，小于则需要收集更多的数据。

由图中，可以看到算法有四类，分类，回归，聚类，降维。

其中 分类和回归是监督式学习，即每个数据对应一个 label。 聚类 是非监督式学习，即没有 label。 另外一类是 降维，当数据集有很多很多属性的时候，可以通过 降维 算法把属性归纳起来。例如 20 个属性只变成 2 个，注意，这不是挑出 2 个，而是压缩成为 2 个，它们集合了 20 个属性的所有特征，相当于把重要的信息提取的更好，不重要的信息就不要了。

然后看问题属于哪一类问题，是分类还是回归，还是聚类，就选择相应的算法。 当然还要考虑数据的大小，例如 100K 是一个阈值。

```python
#### 数据标准化对机器学习成效的影响 
from sklearn import preprocessing 
import numpy as np

# 将资料分割成train与test的模块
from sklearn.model_selection import train_test_split


# 生成适合做classification资料的模块
from sklearn.datasets.samples_generator import make_classification 

# Support Vector Machine中的Support Vector Classifier
from sklearn.svm import SVC 

import matplotlib.pyplot as plt 

#生成具有2种属性的300个数据
X, y = make_classification(
    n_samples=300, n_features=2,
    n_redundant=0, n_informative=2, 
    random_state=22, n_clusters_per_class=1, 
    scale=100)
#可视化数据
plt.scatter(X[:, 0], X[:, 1],c=y)
plt.show()
```

```python
#数据标准化前
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = SVC()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
#数据标准化后
#数据的单位发生了变化, X 数据也被压缩到差不多大小范围.
#标准化后的预测准确率提升至0.9
X = preprocessing.scale(X)
X

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = SVC()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

## 模型评价

scikit-learn模块中的metrics.confusion_matrix函数可以很方便地帮我们计算出混淆矩阵。

要了解各个评价指标，首先需要知道混淆矩阵

- 真正例（TP True Positive），将正类预测为正类数，真实为0，预测也为0。即实际上是正例的数据点被模型标记为正例的数量。
- 假正例（FP False Positive），将负类预测为正类数， 真实为1，预测为0。即实际上是反例的数据点被模型标记为正例的数量。
- 真反例（TP False Negative），将负类预测为负类数， 真实为1，预测为1,即实际上是反例的数据点被模型标记为反例的数量。
- 假反例（FN True Negative），将正类预测为负类数，真实为0，预测为1。即实际上是正例的数据点被模型标记为反例的数量。

```python
#sklearn.metrics.confusion_matrix(y_true, y_pred, labels=None, sample_weight=None)

from sklearn import datasets
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier


#加载iris数据集
iris = datasets.load_iris()
X = iris.data
y = iris.target

#分割数据并
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3)


model = KNeighborsClassifier()
model.fit(X_train, y_train)

print(confusion_matrix(y_test, model.predict(X_test)))
# print(classification_report(y_test, model.predict(X_test)))
```


混淆矩阵可以快速让我们对模型的表现有直观的掌握，但其缺点在于，当面对大量数据场景时候，就不容易直观得衡量模型的优劣。因此在混淆矩阵统计结果的基础上，延伸了4个指标，帮助我们更全面地衡量出模型的性能。

- 准确率（Accuracy Rate），分类模型针对所有实例时，返回正确判断的能力。
- 精确率（Precision Rate），分类模型针对所有相关实例时，返回正确判断的能力。
- 召回率（Recall Rate），分类模型识别所有相关实例的能力。
- F1分数（F1 score），使用调和平均结合召回率和精度的指标。

- 准确率
    是机器学习任务中最常见也和重要的评价指标，是一个可以很直观并且非常容易理解的评价指标，其原理就是被分对的样本数除以所有的样本数。
    
    通常意义上，准确率越高，分类器的性能就越好。但有时候高准确率并不能代表模型就是优秀的，尤其是在样本正负样本存在严重不均衡的情况时，准确率这个评价指标存在着很大的缺陷。例如对某地区在某天是否会发生地震建立预测模型，类别只有发生地震或不发生地震。一个将每个样本都分类为不发生地震的分类器，将拥有超过99%的准确率。可当地震发生时，这个分类器却毫无察觉，这样的分类器无疑会带来了极大的损失。同样的情况还有互联网广告点击预测、互联网金融行业对用户的信用贷款是否逾期预测等等。因此，单纯靠准确率来评价一个模型是远远不全面的。
$$ACC=TP/(TP+TN+FP+FN)$$


- 精确率/查准率
    精确率也被叫查准率，或精度。它是对模型精准性的衡量。其定义为：表示被分为正例的示例中实际为正例的比例，可以简单理解为表示在所有预测结果为辍学的样本中，被正确预测为辍学的比例。其计算公式为：
$$Precision=TP / TP+FP$$

- 召回率/查全率
    召回率通常也被叫做查全率，或灵敏度，它是对模型性能覆盖度的衡量。可以简单理解为表示在所有实际值为辍学的样本中，被正确预测为辍学的比例。其计算公式为：$$Recall=TP/(TP+FN)$$


- F1 Score
    F1分数（F1 Score），又被称为平衡F分数（balanced F Score），用来衡量二分类模型精确度的一种指标，被定义为精确率和召回率的调和平均数，兼顾了分类模型的查准率和查全率。F1的最大值是1，最小值是0。F1分数的值接近查准率和召回率的较小值。
    F1分数的产生，是因为在统计学中，查准率和召回率往往不容易同时考虑，有时分类器可能具有高查准率低召回率或高召回率低查准率。F1分数同时兼顾了分类模型的精确率和召回率。F1分数可以看作是模型精确率和召回率的一种调和平均。
$$F_1=2×(precision ×recall)/(precision+recall)$$

### Cross Validation

Sklearn 中的 Cross Validation (交叉验证)对于我们选择正确的 Model 和 Model 的参数是非常有帮助的， 有了他的帮助，我们能直观的看出不同 Model 或者参数对结构准确度的影响。

```python
from IPython.display import Image as ImageOpen
ImageOpen(filename = "./cross validation.jpg")
```

```python
#普通训练预测
from sklearn.datasets import load_iris # iris数据集
from sklearn.model_selection import train_test_split # 分割数据模块
from sklearn.neighbors import KNeighborsClassifier # K最近邻(kNN，k-NearestNeighbor)分类算法


#加载iris数据集
iris = load_iris()
X = iris.data
y = iris.target

#分割数据并
X_train, X_test, y_train, y_test = train_test_split(X, y)

#建立模型
knn = KNeighborsClassifier()

#训练模型
knn.fit(X_train, y_train)

#将准确率打印出
print(knn.score(X_test, y_test))
```

```python
##random_state
for random_state in range(2,40):
    #分割数据并
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random_state)

    #建立模型
    knn = KNeighborsClassifier()

    #训练模型
    knn.fit(X_train, y_train)

    #将准确率打印出
    print(random_state,knn.score(X_test, y_test))
```

```python
#Model 交叉验证法(Cross Validation) 
from sklearn.cross_validation import cross_val_score # K折交叉验证模块

#使用K折交叉验证模块
scores = cross_val_score(knn, X, y, cv=5, scoring='accuracy')

#将预测准确平均率打印
print(scores.mean())
```

```python
#Model 交叉验证法(Cross Validation) 
from sklearn.cross_validation import cross_val_score # K折交叉验证模块

for cv in range(2,30):
#使用K折交叉验证模块
    scores = cross_val_score(knn, X, y, cv=cv, scoring='accuracy')
    #将预测准确平均率打印
    print(cv,scores.mean())
```

```python
#参数与交叉验证同时进行
import matplotlib.pyplot as plt

#建立测试参数集
k_range = range(1, 31)
k_scores = []

#藉由迭代的方式来计算不同参数对模型的影响，并返回交叉验证后的平均准确率
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=10, scoring='accuracy')
    k_scores.append(scores.mean())
    print(k,scores)


#可视化数据
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
#以平均方差(Mean squared error) 
import matplotlib.pyplot as plt
k_range = range(1, 31)
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    loss = -cross_val_score(knn, X, y, cv=10, scoring='mean_squared_error')
    k_scores.append(loss.mean())

plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated MSE')
plt.show()

#平均方差越低越好，因此选择18左右的K值会最好。
```

```python
from sklearn.learning_curve import learning_curve #学习曲线模块
from sklearn.datasets import load_digits #digits数据集
from sklearn.svm import SVC #Support Vector Classifier
import matplotlib.pyplot as plt #可视化模块
import numpy as np

digits = load_digits()
X = digits.data
y = digits.target
```

```python
#观察样本由小到大的学习曲线变化, 采用K折交叉验证cv=10, 选择平均方差检视模型效能 scoring='mean_squared_error', 样本由小到大分成5轮检视学习曲线(10%, 25%, 50%, 75%, 100%):
train_sizes, train_loss, test_loss = learning_curve(\
                                                    SVC(gamma=0.001), X, y, cv=10, scoring='mean_squared_error',\
                                                    train_sizes=[0.1, 0.25, 0.5, 0.75, 0.80])
#平均每一轮所得到的平均方差(共5轮，分别为样本10%、25%、50%、75%、100%)
train_loss_mean = -np.mean(train_loss, axis=1)
test_loss_mean = -np.mean(test_loss, axis=1)

plt.plot(train_sizes, train_loss_mean, 'o-', color="r", label="train_loss")
plt.plot(train_sizes, test_loss_mean, 'o-', color="g",  label="test_loss")

plt.xlabel("Training examples")
plt.ylabel("Loss")
plt.legend(loc="best")
plt.show()
```

```python
#overfitting导致的loss在训练集上不收敛
train_sizes, train_loss, test_loss = learning_curve(
    SVC(gamma=0.01), X, y, cv=10, scoring='mean_squared_error',
    train_sizes=[0.1, 0.25, 0.5, 0.75, 0.80])

#平均每一轮所得到的平均方差(共5轮，分别为样本10%、25%、50%、75%、80%)
train_loss_mean = -np.mean(train_loss, axis=1)
test_loss_mean = -np.mean(test_loss, axis=1)

plt.plot(train_sizes, train_loss_mean, 'o-', color="r", label="train_loss")
plt.plot(train_sizes, test_loss_mean, 'o-', color="g",  label="test_loss")

plt.xlabel("Training examples")
plt.ylabel("Loss")
plt.legend(loc="best")
plt.show()
```
