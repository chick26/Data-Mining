```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import pandas as pd
```

# 数据无量纲化

## 归一化

```python
data=[[-1,2],[-0.5,6],[0,10],[1,18]]
pd.DataFrame(data)

#实现归一化
scaler=MinMaxScaler() #实例化
scaler=scaler.fit(data)#fit,在这里本质是生成min(x)和max(x)
result=scaler.transform(data)#通过接口导出结果
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(result)
```

```python
pd.DataFrame(scaler.inverse_transform(result))#将归一化后的结果逆转
```

## 标准化

```python
data=[[-1,2],[-0.5,6],[0,10],[1,18]]
pd.DataFrame(data)

#实现归一化
scaler=StandardScaler() #实例化
scaler=scaler.fit(data)#fit,在这里本质是生成min(x)和max(x)
result=scaler.transform(data)#通过接口导出结果
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(result)
```

```python
pd.DataFrame(scaler.inverse_transform(result))#将归一化后的结果逆转
```

# 缺失值处理

## sklearn.simpleImputer

```python
data = pd.read_csv('./data/Narrativedata.csv',index_col=0)
data.head()
```

```python
data.info()
```

```python
Age = data.loc[:,'Age'].values.reshape(-1, 1) # sklearn 处理特征必须是二维的
Age[:20]
```

```python
imp_mean = SimpleImputer()
imp_median = SimpleImputer(strategy='median')
imp_0 = SimpleImputer(strategy='constant', fill_value=0)
```

```python
# imp_mean_model = imp_mean.fit(Age)
# imp_mean_result = imp_mean_model.transform(Age)
imp_mean_result = imp_mean.fit_transform(Age)
imp_mean_result
```

```python
imp_median_result = imp_median.fit_transform(Age)
imp_median_result
```

```python
imp_0_result = imp_0.fit_transform(Age)
imp_0_result
```

```python
# 使用中位数填充 Age
data.loc[:,"Age"]=imp_median
data.info()
```

```python
#使用众数填补Embarked,众数可以填充文字型数据
Embarked=data.loc[:,"Embarked"].values.reshape(-1,1)
imp_mode=SimpleImputer(strategy="most_frequent")
data.loc[:,"Embarked"]=imp_mode.fit_transform(Embarked)
data.info()
```

## 处理分类型特征：编码与哑变量

### 标签

```python
from sklearn.preprocessing import LabelEncoder#标签专用

data = pd.read_csv('./data/Narrativedata.csv',index_col=0)

#index_col意思是把第0列作为索引
y=data.iloc[:,-1]#要输入的是标签，不是特征矩阵，所有允许一维（一般都习惯放到最后一列，这里是survived  ）

le=LabelEncoder()#实例化
le=le.fit(y)#导入数据
label=le.transform(y)#transfrom接口调取结果

# data.iloc[:,-1]=LabelEncoder().fit_transform(data.iloc[:,-1])
```

```python
le.classes_ # 查看标签类别
```

```python
data.iloc[:,-1]=label#让标签等于我们运行出来的结果
```

### 特征

```python
from sklearn.preprocessing import OrdinalEncoder

data_ = data.copy()

data_.loc[:,"Age"]=data_.loc[:,"Age"].fillna(data_.loc[:,"Age"].median())
data_.dropna(axis=0,inplace=True)

#需要保证没有缺失值，如果有缺失值，则会报错
OrdinalEncoder().fit(data_.iloc[:,1:-1]).categories_#data_.iloc[:,1:-1]所有的行从索引为1的列取到最后1列（不包括最后1列）
```

```python
data_.iloc[:,1:-1]=OrdinalEncoder().fit_transform(data_.iloc[:,1:-1])
```

```python
data_
```

### 使用独热编码，创建哑变量

```python
from sklearn.preprocessing import OneHotEncoder

X = data.iloc[:,1:-1]
X.dropna(axis=0,inplace=True)
enc = OneHotEncoder(categories='auto').fit(X)
result = enc.transform(X).toarray()#转化为数组
# # enc.transform(X)
result#因为是sex和embarked两个特征，sex有两类，embarked有三类，所以最后结果是5列
```

```python
enc.get_feature_names()
```

```python
new_data =pd.concat([data, pd.DataFrame(result)], axis=1)
new_data.drop(['Sex', 'Embarked'], axis=1, inplace=True)
new_data.columns = ['Age','Survived', 'x0_female', 'x0_male', 'x1_C', 'x1_Q', 'x1_S']
```

### 处理连续型特征，二值化和分段

#### 二值化

```python
from sklearn.preprocessing import Binarizer
```

```python
data_2=data.copy()
data_2.dropna(axis=0,inplace=True)

X = data_2.iloc[:,0].values.reshape(-1,1)#Binarizer类为特征专用，所以不能使用一维数组
transformer=Binarizer(threshold=30).fit_transform(X)#30以上的划为1，以下的划为0
#transformer

data_2.iloc[:,0]=transformer
```

```python
data_2
```

#### 分段

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
data_3=data.copy()
data_3.dropna(axis=0,inplace=True)

X = data_3.iloc[:,0].values.reshape(-1,1)
est = KBinsDiscretizer(n_bins=3,encode='ordinal',strategy='uniform')
est.fit_transform(X)
```
