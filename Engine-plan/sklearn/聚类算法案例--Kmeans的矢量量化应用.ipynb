```python
#现在我们发现，一共有9w多种颜色，我们希望压缩一下，使用Kmeans将颜色压缩到64种，并且不严重损耗图像质量
"""
将64个质心来全部替代9w多种颜色
随机选取64个样本点作为随机质心，计算原数据中每个样本到它们的距离来找出每个样本最近的随机质心
然后用质心替换原来的样本
"""
```

```python
#Kmeans不接受三维数据，所以要变成2维的
#首先做归一化处理，压缩到[0,1]之间（这里的归一化和之前数据归一化用的公式不一样），因为plt.imshow在浮点数下表现优秀
n_clusters=64
#将所有数据都变成【0,1】
china=np.array(china,dtype=np.float64)/china.max()
#把china从图像格式，转换成矩阵格式
w,h,d=original_shape=tuple(china.shape)
```

```python
image_array=np.reshape(china,(w*h,d))#reshape是改变结构的函数
#np.reshape(a,newshape,order='C'),第一个参数a是要改变结构的对象，第二个参数是要改变的新结构。第三个参数不用在意
image_array.shape
```

```python
#开始建模
"""对数据进行Kmeans矢量量化"""
#首先，先使用1000个数据来找质心
image_array_sample=shuffle(image_array,random_state=0)[:1000]
kmeans=KMeans(n_clusters=n_clusters,random_state=0).fit(image_array_sample)
kmeans.cluster_centers_
```

```python
#找出质心后，按照已存在的质心对所有数据进行聚类
labels=kmeans.predict(image_array)
labels.shape
```

```python
#使用质心来替换掉所有的样本
image_kmeans=image_array.copy()
image_kmeans
```

```python
for i in range(w*h):
    image_kmeans[i]=kmeans.cluster_centers_[labels[i]]
```

```python
image_kmeans
```

```python
pd.DataFrame(image_kmeans).drop_duplicates().shape#看看现在替换了质心之后还有多少种颜色
```

```python
#恢复图片结构，否则不能imshow
image_kmeans=image_kmeans.reshape(w,h,d)
image_kmeans.shape
```

```python
"""对数据进行随机的矢量量化"""
centroid_random=shuffle(image_array,random_state=0)[:n_clusters]
#现在随机抽取64个点作为质心
labels_random=pairwise_distances_argmin(centroid_random,image_array,axis=0)
#计算这27w个点到64个质心点的距离，然后找到27w个点距离最近的这个质心，作为这个点的索引
```

```python
image_random=image_array.copy()
for i in range(w*h):
    image_random[i]=centroid_random[labels_random[i]]
image_random=image_random.reshape(w,h,d)
image_random.shape
```

```python
#将原图，对数据进行Kmeans矢量量化，对数据进行随机的矢量量化这三种情况的图像画出来，进行对比
plt.figure(figsize=(10,10))
plt.axis('off')
plt.title('original image (96615 colors)')
plt.imshow(china)

plt.figure(figsize=(10,10))
plt.axis('off')
plt.title('quantized image (64 colors,Kmeans)')
plt.imshow(image_kmeans)


plt.figure(figsize=(10,10))
plt.axis('off')
plt.title('quantized image (64 colors,random)')
plt.imshow(image_random)
plt.show()
```
