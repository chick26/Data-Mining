```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris#鸢尾花数据集
from sklearn.decomposition import PCA
```

```python
iris=load_iris()
y=iris.target
x=iris.data
```

```python
min(x.shape)
```

```python
import pandas as pd
pd.DataFrame(x)#现在是一个4维特征向量的矩阵，想要变成可视化，降维到2维
```

```python
y
```

```python
#调用PCA
pca=PCA(n_components=2)#实例化
pca=pca.fit(x)#拟合模型
x_dr=pca.transform(x)#获取新矩阵
x_dr.shape#实现了降维
```

```python
plt.figure()
#alpha=0.6，alpha是表示透明度
plt.scatter(x_dr[y==0,0],x_dr[y==0,1],c="red",label=iris.target_names[0],alpha=0.6)
plt.scatter(x_dr[y==1,0],x_dr[y==1,1],c="blue",label=iris.target_names[1],alpha=0.6)
plt.scatter(x_dr[y==2,0],x_dr[y==2,1],c="pink",label=iris.target_names[2],alpha=0.6)
plt.legend()
plt.title("PCA of IRIS dataset")
plt.show()
```

```python
#探索降维后的数据
pca.explained_variance_#可见大部分的信息都在第一个特征上
```

```python
pca.explained_variance_ratio_#explained_variance_ratio_是占原始信息的百分比
#第一个特征占了92%
```

```python
pca.explained_variance_ratio_.sum()#表示降维后所有的特征占总信息的百分比
```

```python
#那么现在选取n_components=2的时候发现携带信息还很多，如何选择n_components的取值呢
#当n_components不填写任何值的时候，可以得到累积可解释方差贡献率
```

```python
#即现在这样是4个特征，每个特征占原始信息的百分比
pca_line=PCA().fit(x)
pca_line.explained_variance_ratio_
```

```python
import numpy as np
np.cumsum(pca_line.explained_variance_ratio_)
```

```python
import numpy as np
pca_line=PCA().fit(x)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4])#限制坐标轴显示为整数，如果不写，则显示为1,1.5,2,2.5,3,3.5,4
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance ratio")
plt.show()
```

```python
#可以用最大似然估计自选n_components的值
pca_mle=PCA(n_components="mle")
pca_mle=pca_mle.fit(x)
x_mle=pca_mle.transform(x)
x_mle#此时显示最大似然估计选的是3个特征
```

```python
#还可以按信息量占比选超参数（比如选占原信息90%）
pca_f=PCA(n_components=0.98,svd_solver="full")
pca_f=pca_f.fit(x)
x_f=pca_f.transform(x)
```

```python
pca_f.explained_variance_ratio_
```

```python
axes#axes中的一个对象对应fig中的一个空格
```

```python
#我们使用子图对象.imshow来将图像填充到空白画布上
#而imshow要求的数据格式必须是一个(m,n)格式的矩阵，即每个数据都是一张单独的图
#因此我们需要遍历的是faces.images，其结构是（1348,62,47）
#所以我们要做的是[i,:,:]即，遍历第i行的图像
#例如：
axes[0][0].imshow(faces.images[0,:,:])
#fig中第0行第0列的图像
```

```python
fig
#我们发现第一个图像就已经被填充上了
```

```python
#接下来我们要进行遍历循环
#由于axes是一个4*5的矩阵，要循环很麻烦，我们可以将axes降维，然后进行填充
len([*axes.flat])#这样就将axes转成了一个一维数组
```

```python
[*enumerate(axes.flat)]#加上enumerate之后就在这个一维数组中加上了索引
```

```python
fig,axes=plt.subplots(3,8#生成3*8个格子
                      ,figsize=(8,4)#横向8，纵向4
                      ,subplot_kw={"xticks":[],"yticks":[]}#不要显示坐标轴
                     )
for i,ax in enumerate(axes.flat):
    ax.imshow(faces.images[i,:,:],cmap="gray")#选择色彩模式
```

```python
#接下来进行降维
pca=PCA(150).fit(x)#x是faces.data，将它降维到150
#为什么要用faces.data而不用faces.images呢，因为sklearn中只接受小于等于二维的
```

```python
V=pca.components_
```

```python
V.shape#得到的是V(k,n)
```

```python
#V成原有的矩阵x就是降维后的，这里就是x*VT 得到的是一个1348*150的数据
#将新特征空间矩阵进行可视化
```

```python
fig,axes=plt.subplots(3,8#生成3*8个格子
                      ,figsize=(8,4)#横向8，纵向4
                      ,subplot_kw={"xticks":[],"yticks":[]}#不要显示坐标轴
                     )
for i,ax in enumerate(axes.flat):
    ax.imshow(V[i,:].reshape(62,47),cmap="gray")#选择色彩模式
    #因为此时V是一个150*2914的矩阵，V[0]是（2914,）而V[0].reshape(62,47)就是还原成（62,47）的矩阵
```

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
faces=fetch_lfw_people(min_faces_per_person=60)
faces.images.shape
```

```python
faces.data.shape
```

```python
x=faces.data
```

```python
#建模降维，获取降维后的特征矩阵x_dr
pca=PCA(150)
x_dr=pca.fit_transform(x)
x_dr.shape
```

```python
x_inverse=pca.inverse_transform(x_dr)
x_inverse.shape
```

```python
#接下来对x和x_inverse分别进行可视化，看看是不是一样的
```

```python
fig,ax=plt.subplots(2,10,figsize=(10,2.5)
                   ,subplot_kw={"xticks":[],"yticks":[]} 
                  )
for i in range(10):
    ax[0,i].imshow(faces.images[i,:,:],cmap="binary_r")
    ax[1,i].imshow(x_inverse[i].reshape(62,47),cmap="binary_r") 
    #得到的数据下面的比上面的要模糊。这代表经过降维后损失的信息，虽然又经过inverse但是得到的依然不是原来的信息了
```

```python
#PCA做噪音过滤
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
digits=load_digits()
digits.data.shape
```

```python
#定义画图函数
def plot_digits(data):
    fig,axes=plt.subplots(4,10,figsize=(10,4)
                         ,subplot_kw={"xticks":[],"yticks":[]}
                         )
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),cmap="binary")
plot_digits(digits.data)#没有噪声的样子
```

```python
#为数据加上噪音
import numpy as np
rng=np.random.RandomState(42)
#在指定数据集中，随机抽取服从正态分布的数据
#两个参数，分别是指定的数据集，和抽取出来的正态分布方差
noisy=rng.normal(digits.data,2)
plot_digits(noisy)
```

```python
#接下来对noisy进行降噪
```

```python
#先进行降维
pca=PCA(0.5,svd_solver="full").fit(noisy)
x_dr=pca.transform(noisy)
#先降维的目的是除掉大部分的噪音
```

```python
x_dr.shape
```

```python
#逆转降维结果，实现降噪
without_noise=pca.inverse_transform(x_dr)
plot_digits(without_noise)
```
