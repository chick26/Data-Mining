```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
#test_size=0.3代表30%做测试集，70%做训练集
Xtrain,Xtest,Ytrain,Ytest=train_test_split(wine.data,wine.target,test_size=0.3)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

```python
#sklearn 建模三部曲
clf=tree.DecisionTreeClassifier(criterion="entropy"
                                ,random_state=30
                                ,splitter="random"
#                                 ,min_samples_leaf=10
#                                 ,min_samples_split=10
#                                 ,max_depth=3
                               )
#加上random_state（多少都行）之后score的结果就不再变化
#random_state和splitter都是防止过拟合
clf=clf.fit(Xtrain,Ytrain)
score=clf.score(Xtest,Ytest)#返回预测的准确度accuracy
score#分数越高越好。这个分数和过拟合没关系
```

```python
feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类','花青素','颜色强度','色调','od280/od315稀释葡萄酒','脯氨酸']
import graphviz
dot_data=tree.export_graphviz(clf
                              ,feature_names=feature_name
                              ,class_names=["啤酒","白酒","红酒"]
                              ,filled=True
                              ,rounded=True               
                             )
graph=graphviz.Source(dot_data)
graph
#图像中的颜色不一样（不纯度entropy越低对应的颜色越深）filled=True这句代表填充颜色
#rounded=True代表圆角矩形，若去掉则变成矩形
```

```python
clf.feature_importances_
#代表特征的重要性，重要性越大的值越大
```

```python
[*zip(feature_name,clf.feature_importances_)]
#显示脯氨酸的权重最大（根节点的权重永远都是最大的）
```

```python
clf=tree.DecisionTreeClassifier(criterion="entropy"
                                ,random_state=30
                                ,splitter="random"
#                                 ,min_samples_leaf=10
#                                 ,min_samples_split=10
                                ,max_depth=3
                               )
#加上random_state（多少都行）之后score的结果就不再变化
#random_state和splitter都是防止过拟合
clf=clf.fit(Xtrain,Ytrain)
dot_data=tree.export_graphviz(clf
                              ,feature_names=feature_name
                              ,class_names=["啤酒","白酒","红酒"]
                              ,filled=True
                              ,rounded=True               
                             )
graph=graphviz.Source(dot_data)
graph
```

```python
import matplotlib.pyplot as plt
test = []
for i in range(10):
    clf=tree.DecisionTreeClassifier(criterion="entropy"
                                    ,random_state=30
                                    ,splitter="random"
    #                                 ,min_samples_leaf=10
    #                                 ,min_samples_split=10
                                    ,max_depth=i+1
                                   )
    clf = clf.fit(Xtrain,Ytrain)
    score = clf.score(Xtest,Ytest)
    test.append(score)
plt.plot(range(1,11),test,color="red",label="max_depth")
plt.legend()
plt.show()
```

```python
#apply返回每个测试样本所在的叶子节点的索引
clf.apply(Xtest)
```

```python
#predict返回每个测试样本的分类/回归结果
clf.predict(Xtest)
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
rng=np.random.RandomState(1)#随机种子
#和树中的random_state一样 
```

```python
X=np.sort(5*rng.rand(80,1),axis=0)
#生成一个0-5的二维随机数组，然后从小到大进行排序
y=np.sin(X).ravel()#生成正弦曲线
#用ravel降维成一维
y[::5]+=3*(0.5-rng.rand(16))
#[::5]代表每5个取出来一个，即y中的80个数字每5个取出来一个随机
#即这一步是在添加噪音，模拟现实中的噪音点，让模型变得普通一些
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="blue",c="orange", label="data")
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X, y)
regr_2.fit(X, y)
```

```python
#np.random.rand(数组结构)，生成随机数组的函数#了解降维函数ravel()的用法
#ravel()可以将n维降到n-1维
#np.random.random((2,1))
```

```python
#np.random.random((2,1)).ravel()
```

```python
#np.random.random((2,1)).ravel().shape
```

```python
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
#np.arange(开始点，结束点，步长)
#结果是[开始点，结束点)左闭右开
#[:, np.newaxis]用来增维，将X_test可以放入fit中
```

```python
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
#利用max_depth不同分别得到不同的回归结果
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="red",c="blue", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
#由下图可以看到绿色的曲线对训练集拟合程度高（过拟合），容易受到噪音的影响
#蓝色的曲线表现好于绿色
```
