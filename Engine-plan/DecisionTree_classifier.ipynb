```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn import tree
from sklearn.model_selection import train_test_split
```

## 导入数据

```python
wine = load_wine()
# wine.data.shape
```

```python
pd.concat([pd.DataFrame(wine.data), pd.DataFrame(wine.target)], axis=1)
```

## 拆分数据集

```python
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size = 0.3)
```

## 建立模型

```python
'''
criterion 不纯度算法： entropy/gini
splitter 控制决策树中的随机选项，”best"，优先选择更重要（按照不纯度设置的基尼系数或者信息熵的值）“random"，决策树在分枝时会更加随机
random_state  随机选择节点特征(类似 random_seed)
max_depth 决策树深度
min_samples_split  最小节点分支
min_samples_leaf 限制叶子节点样本数（限制过拟合)
max_feature
min_impurity_decrease
class_weight 平衡样本
min_weight_fraction_leaf

'''
estimator = tree.DecisionTreeClassifier(
    criterion='entropy', 
    random_state=100,
    splitter='best',
)
model = estimator.fit(X_train,y_train)
```

```python
score = model.score(X_test, y_test)
score
```

## 可视化决策树

```python
import graphviz
feature_name = ['酒精', '苹果酸', '灰', '灰的碱性', '镁', '总酚', '类黄酮',
                '非黄烷类酚类', '花青素', '颜色强度', '色调', 'od280/od315稀释葡萄酒', '脯氨酸']

dot_data = tree.export_graphviz(
    model,
    out_file=None,
    feature_names=feature_name,
    class_names=["琴酒","雪莉","贝尔摩德"],
    filled=True,
    rounded=True
)
graph = graphviz.Source(dot_data)
graph
```

## 探索决策树

## 通过学习曲线确定最优参数

```python
import matplotlib.pyplot as plt
```

```python
scores = []
for i in range(10):
    estimator = tree.DecisionTreeClassifier(max_depth=i+1, random_state = 100)
    model = estimator.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    scores.append(score)
    
plt.figure(figsize=[25,5])
plt.plot(range(1,11), scores)
```

```python
max_score = max(scores)
best_index = scores.index(max_score)
print('最高得分 %.2f' % max_score)
print('最佳深度 %d' % (best_index+1))
```
