# 红酒数据随机森林

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
```

## 数据导入

```python
wine = load_wine()
```

## 数据建模

```python
%%time

rfc = RandomForestClassifier(n_estimators=25, random_state=100)
dfc = DecisionTreeClassifier(random_state=100)
scores_rfc = cross_val_score(rfc , wine.data, wine.target, cv=10)
scores_dtc = cross_val_score(dfc , wine.data, wine.target, cv=10) 
```

## 模型对比 - 1

```python
plt.figure(figsize=[25,5])
plt.title('RandomForestClassifier VS DecisionTreeClassifier')
plt.plot(range(1, 11), scores_rfc, label='RandomForestClassifier')
plt.plot(range(1, 11), scores_dtc, label='DecisionTreeClassifier')
plt.legend()
```

## 模型对比 - 2

```python
scores_rfc = []
scores_dtc = []
for i in range(1,11):
    rfc = RandomForestClassifier(n_estimators=25)
    dfc = DecisionTreeClassifier()
    scores_rfc.append(cross_val_score(rfc , wine.data, wine.target, cv=5).mean())
    scores_dtc.append(cross_val_score(dfc , wine.data, wine.target, cv=5).mean())

plt.figure(figsize=[25,5])
plt.title('RandomForestClassifier VS DecisionTreeClassifier')
plt.plot(range(1, 11), scores_rfc, label='RandomForestClassifier')
plt.plot(range(1, 11), scores_dtc, label='DecisionTreeClassifier')
plt.legend()
```
